```
import pandas as pd
import numpy as np
```

```
data = pd.read_csv('../data/boston.csv', index_col=0)
data
```

```
#1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?
data.boxplot(figsize=(8, 6))
```

```
# 驗證箱型圖
data.describe()
```

```
a = data.median()[300<=data.median()][data.median()<=400]
print(a)
print('='*50)
a.index
```
