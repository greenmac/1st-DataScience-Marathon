矩陣相乘維度需要對好，例如:2X3矩陣乘上3X5矩陣得到2X5矩陣

題目:<br>
array1 = np.array([[10, 8], [3, 5]]) <br>
1. 運用上列array計算反矩陣，乘上原矩陣，並觀察是否為單位矩陣?
2. 運用上列array計算特徵值、特徵向量?
3. 運用上列array計算SVD?

```
import numpy as np
```

```
#1. 運用上列array計算反矩陣，乘上原矩陣，並觀察是否為單位矩陣?
array1 = np.array([[10, 8], [3, 5]])

inv_array1 = np.linalg.inv(array1)
print(inv_array1) # 利用反矩陣證明是否為單位矩陣
print(np.matmul(array1, inv_array1)) # 利用 matmul 鄭兩市矩陣相乘, 不是矩陣*純向量
```

```
#2. 運用上列array計算特徵值、特徵向量?
print('eigen value:', np.linalg.eig(array1)[0])
print('eigen vector:', np.linalg.eig(array1)[1])
```

```
#3. 運用上列array計算SVD?
print(np.linalg.svd(array1))
```
